## Investigating missing data
There should be `38 * 20 * 2 = 1520` games

Checking the `value_counts` for each team:

```python
matches["team"].value_counts()
```

Each team should have `38 * 2` games other than the relegated/promoted teams.

It is clear that Liverpool are missing matches.

```python
matches[matches["team"] == "Liverpool"]
```

A whole season is missing for Liverpool

```python
matches["round"].value_counts()
```

There should be 39 (excluding Liverpool games) but we are missing same game weeks caused by the data being scraped prior to these weeks occurring

```python
matches.dtypes
```

Machine learning algorithms can only work with numerical data, not objects. This means we have to convert the columns of object type to int/float.

Convert date column from object to datetime:

```python
matches["date"] = pd.to_datetime(matches["date"])
```

Convert venue column from object to categorical then assigning a numerical value using `cat.codes`

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

Create a unique code for each opponent:

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

Take the time column and only take the hour that the game occurred (using a regular expression):

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

Get the day of the week the game occurred:

```python
matches["day_code"] = matches["date"].dt.dayofweek
```
